**global variables and local variables**

global mean we can access from anywhere
local mean we can access only from the function

```python
g=500   ### "global variable"

def data_science():
  print(f'iam calling from inide the function: {g}')

data_science() #calling the function

print(f'iam calling from outside the function: {g}')
```

```python
def data_science():    ## local variable
  h=200
  print(f'iam calling from inide the function: {h}')

data_science() #calling the function

print(f'iam calling from outside the function: {h}')
```
