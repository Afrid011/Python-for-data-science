
-type conversion-

converting one data type into another

- 'int->float and int->str
- 'float->int and float->str
- 'str->int and str->float

```python
# int -> float

a=25 # int will save in the memory
print(a) # calling from memory
print(type(a)) # checking data type

b=float(a) #conversion is happening

print(b) #calling from memory
print(type(b)) #calling and checking data type
```

```python
# int -> str

a=25 # int will save in the memory
print(a) # calling from memory
print(type(a)) # checking data type

b=str(a) #conversion is happening

print(b) #calling from memory
print(type(b)) #calling and checking data type
```
