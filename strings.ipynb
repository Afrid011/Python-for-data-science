**Mututable or Immutable**

mutable ->> if any concept after saving into memory if we change the memory size we will  call them as mutable

immutable ->> if any concept after saving into memory uf we cant change their memory size we call them as immutable

```python
# trying yo add the extra memory to the existing memory


a='data' # will save in the memory

a[4]='s' # here trying to create 4th index and wanto fill the character J in it

print(a)
```

```python
 # trying to decease the memory size

 a='data' # will save in memory
 del(a[2])
 print(a)
```

**therefore the strings are immutable
once the string is assigned into the memory we cant decrease tor increase the size of the memory**
