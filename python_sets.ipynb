```python
# add ->> add function will take a single value from the user and
# will add to the existing set in some random location

a={10,20,30,40,50} # will save in the memory
print(f'the values in a is {a}')

a.add(60) # somewhere randomly 60 will be saved

print(f'after adding values in a is {a}')
```

```python
# update ->> will take a group of values which s neded to be written in list and those will be updates into the set

a={10,20,30,40,50} # will save in the memory
print(f'the values in a is {a}')

a.update([1,2]) # somewhere randomly will be saved

print(f'after adding values in a is {a}')
```

```python
# for deleting
# pop ->> it will simply remove only one value from the set

a={10,20,30,40,50} # will save in the memory
print(f'the values in a is {a}')

a.pop() # remove one value randomly

print(f'after adding values in a is {a}')
```

```python
# discard ->> will take the valuve from the used and if the value is there then it will remove or the value is not present then no work happens

a={10,20,30,40,50} # will save in the memory
print(f'the values in a is {a}')


a.discard(40) # remove the vlue 40 from the memory
print(f'after dicarding the value in a is {a}')
```

```python
# remove ->> will take the valuve from the used and if the value is there then it will remove or the value is not present then no work happens

a={10,20,30,40,50} # will save in the memory
print(f'the values in a is {a}')


a.remove(40) # remove the vlue 40 from the memory
print(f'after dicarding the value in a is {a}')
```

```python
# copy ->> is used to copy the data from one variable to another varible where it makes variables to have different address
# deep copy ->> have same addresses
# shallow copy ->> have different ddresses
a={10,20,30,40,50} # will save in the memory

b=a
print(a)
print(b)

a.add(100)

print(a)
print(b)

print(id(a))
print(id(b))
```

```python
# clear ->> it will remove all the data from the variable but variable will be alive

a={10,20,30,40,50} # will save in the memory

print(a)

a.clear()

print(a)
```

```python
# union

a={10,20,30,40,50} # will save in the memory
b={1,2,3,4,5}

print(a.union(b))
```

```python
# intersection

a={10,20,30,40,50} # will save in the memory
b={1,2,3,4,5}

print(a.intersection(b))
```
