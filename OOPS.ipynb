**Inheritance**

inheritance ->> parent nd child relationship

there are 5 types of inheritance

-  single inheritance
-  multiple inheritance
-  multilevel  inheritance
-  hirarchial inheritance
-  hybrid inheritance

**single inheritance**

```python
class PARENT:
  def building(self):
    print(f"i have a building in chennai")


class CHILD(PARENT):
  def job(self):
    print(f"im doing my job")


anil=CHILD()
anil.job()
anil.building() #single inheritance
```

**multilevel inheritance**

```python
class PARENT:
  def building(self):
    print(f"i have a building in chennai")


class CHILD(PARENT):
  def job(self):
    print(f"im doing my job")


class GRANDCHILD(CHILD,PARENT):
  def money(self):
    print(f'i got my pocket money')

kiran=CHILD()
kiran.job()
kiran.building()


anil=GRANDCHILD()
anil.money()
anil.job()
anil.building()
```

**multiple inheritance**

```python
class FATHER:
  def building(self):
    print(f"i have a building in chennai")


class MOTHER:
  def job(self):
    print(f"im doing my job")


class CHILD(FATHER,MOTHER):
  def money(self):
    print(f'i got my pocket money')


kumar=CHILD()
kumar.job()
kumar.building()
kumar.money()
```

**hybrid inheritance**

```python
class GRANDFATHER:
  def building(self):
    print(f"i have a building in chennai")


class FATHER(GRANDFATHER):
  def job(self):
    print(f"im doing my job")


class MOTHER(GRANDFATHER):
  def money(self):
    print(f'i got my pocket money')

class CHILD(FATHER,MOTHER):
  def store(self):
    print(f'i have my own store')


shiv=CHILD()
shiv.job()
shiv.building()
shiv.money()
shiv.store()
```
