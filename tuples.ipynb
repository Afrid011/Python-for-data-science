-  tuple is repesented with ->> () \
-  tuple can accept any data type -> int | float | str
-  tuple is the smallest concept in the entire python
-  for a single variable we can pass multiple inputs

```python
a=(10,55.55,'ass123') # will save  memory

print(a)
print(a[1])
```

```python
a=(2,3,4,4,3,2,3,4,4,3,2,3,4,4,3,2,3,2)
print(a)
print(a.count(3))
print(a.index(4))
print(a[::-1])
```

```python
a=(1,2,3,4,5,6,7,8,9)
print(a[2::5])
```

**tuple in python is mutuable or immutable**
-  addition
-  deletion

```python
#add

a=(10,14,15,9,4,20) #will save in the memory
a[7]=8
print(a)
```

```python
#delete
a=(10,14,15,9,4,20) #will save in the memory

del a[2]
print(a)
```

**BUILT IN FUNCTIONS**

-  index

```python
# count

a=(10,14,15,9,4,20)
print(a.count(10))
```

```python
#index

a=(10,14,15,9,4,20)
print(a.index(15))
```
