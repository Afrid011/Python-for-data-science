**Multiprocesing**

```python
import multiprocessing
```

```python
def india():
  for i in range(1,11):
    time.sleep(0.5)
    print('india->',i)

def usa():
  for i in range(51,61):
    time.sleep(0.5)
    print('usa->',i)


if __name__=='__main__':
  starting_time=time.time()

  c1=multiprocessing.Process(target=india,args=())
  c2=multiprocessing.Process(target=usa,args=())

  c1.start()
  c2.start()

  c1.join()
  c2.join()

  ending_time=time.time()
  print(ending_time-starting_time)
```

```python
#multiprocesing ->> creating multiple processors and working on it

import multiprocessing
import time
import os

def india():
  print(f'c1 procesor id : {os.getpid()}')
  for i in range(1,11):
    time.sleep(0.5)
    print('india->',i)

def usa():
  print(f'c2 procesor id : {os.getpid()}')
  for i in range(51,61):
    time.sleep(0.5)
    print('usa->',i)


if __name__=="__main__":

    starting_time=time.time()
    print(f'mainprocessor is started:{os.getpid()}')
    c1=multiprocessing.Process(target=india,args=())
    c2=multiprocessing.Process(target=usa,args=())

    c1.start()
    c2.start()

    print(f'---------main processor exists-------')
    print(f'total time : {time.time-starting_time}')

```

```python
import multiprocessing
import time
import os
import threading


def india():
  print(f'c1 procesor id : {os.getpid()}')
  for i in range(1,11):
    time.sleep(0.5)
    print('india->',i)

def canada():
  print(f'c1 procesor id : {os.getpid()}')
  for i in range(1,11):
    time.sleep(0.5)
    print('canada->',i)

def c1_common():
  t1=threading.Thread(target=india,args=())
  t2=threading.Thread(target=canada,args=())
  t1.start()
  t2.start()


if __name__=="__main__":

    starting_time=time.time()
    print(f'mainprocessor is started:{os.getpid()}')
    c1=multiprocessing.Process(target=c1_common,args=())


    c1.start()


    c1.join()

    ending=time.time()
    print(f'total time: {ending-starting_time}')
    print(f'---------main processor exists-------')
```
